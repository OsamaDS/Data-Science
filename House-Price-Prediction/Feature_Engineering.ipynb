```python
import pandas as pd
import numpy as np
pd.pandas.set_option('display.max_columns',None)
```

# Missing Values

```python
cat_nan_feature=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == "O"]
for feature in cat_nan_feature:
    print('{} {}'.format(feature,dataset[feature].isnull().sum()))
```

```python
for feature in cat_nan_feature:
    dataset[feature]=dataset[feature].replace(np.nan,'Missing')
    print('{} {}'.format(feature,dataset[feature].isnull().sum()))     
```

```python
num_nan_feature=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != "O"]
for feature in num_nan_feature:
    print('{} {}'.format(feature,dataset[feature].isnull().sum()))
```

```python
for feature in num_nan_feature:
    feat_median=dataset[feature].median()
    dataset[feature]=dataset[feature].replace(np.nan,feat_median)
    print('{} {} and median is {}'.format(feature,dataset[feature].isnull().sum(),feat_median)) 
```

# change numerical features which are skwed, into standard normal distribution

```python
num_feature=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])
dataset[num_feature].head()
```

# Removing 1% catagories of categorical features

```python
cat_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=="O"]
cat_feature
```

```python
for feature in cat_feature:
    tmp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    tmp_ind=tmp[tmp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(tmp_ind),dataset[feature],"Rare_var")
#tmp[tmp>0.01].index  
```

```python
dataset[cat_feature][50:100]
```

```python
for feature in cat_feature:
    label=dataset.groupby(feature)["SalePrice"].mean().sort_values().index
    label={n:m for m,n in enumerate(label,0)}
    #print(label)
    dataset[feature]=dataset[feature].map(label)
dataset[cat_feature].head()
```

# Scaling

```python
scale_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[scale_feature])
```

```python
data=pd.concat([dataset[['Id','SalePrice']],pd.DataFrame(scaler.transform(dataset[scale_feature]),columns=scale_feature)],axis=1)
```
